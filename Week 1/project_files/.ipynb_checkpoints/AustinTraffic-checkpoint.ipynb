```python
# Extract hour from datetime
df['Hour'] = df['Published Date'].dt.hour

# Define rush hour times (7–9am and 4–6pm)
def is_rush_hour(hour):
    return (7 <= hour <= 9) or (16 <= hour <= 18)

df['Rush Hour'] = df['Hour'].apply(is_rush_hour)

# Preview
df[['Published Date', 'Hour', 'Rush Hour']].head()
```

```python
import matplotlib.pyplot as plt

# Inline plotting
%matplotlib inline

plt.figure(figsize=(10, 6))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.3, s=1)
plt.title("Traffic Incidents in Austin")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```

```python
austin_bounds = {
    'lat_min': 30.10,
    'lat_max': 30.60,
    'lon_min': -97.95,
    'lon_max': -97.40
}

df = df[
    (df['Latitude']  >= austin_bounds['lat_min']) &
    (df['Latitude']  <= austin_bounds['lat_max']) &
    (df['Longitude'] >= austin_bounds['lon_min']) &
    (df['Longitude'] <= austin_bounds['lon_max'])
]
```

```python
df
```

```python
# Get the top 5 issue types
top5 = df['Issue Reported'].value_counts().nlargest(5).index.tolist()

# Define colors for each
color_map = {
    issue: color for issue, color in zip(top5, ['red', 'blue', 'green', 'orange', 'purple'])
}

fig, ax = plt.subplots(figsize=(10, 8))

# Plot everything else first as a grey background
df_other = df[~df['Issue Reported'].isin(top5)]
ax.scatter(
    df_other['Longitude'], df_other['Latitude'],
    c='lightgrey', s=5, alpha=0.3, label='Other'
)

# Plot each top 5 issue type with its own color
for issue in top5:
    df_issue = df[df['Issue Reported'] == issue]
    ax.scatter(
        df_issue['Longitude'], df_issue['Latitude'],
        c=color_map[issue], s=10, alpha=0.5, label=issue
    )

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Austin Traffic Incidents — Top 5 Issue Types')
ax.legend(markerscale=2, loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.hist(df['Hour'], bins=24, color='orange', edgecolor='black')
plt.title("Incidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Number of Incidents")
plt.grid(True)
plt.show()
```

```python
plt.hist(df['Month'], bins=12, color='purple', edgecolor='black')
plt.title("Incidents by Month")
plt.xlabel("Month")
plt.ylabel("Number of Incidents")
plt.grid(False)
plt.show()
```

```python
top_issues = df['Issue Reported'].value_counts().head(5)
top_issues.plot(kind='bar', color='skyblue')
plt.title("Top 5 Traffic Issues Reported")
plt.xlabel("Issue")
plt.ylabel("Number of Reports")
plt.xticks(rotation=45)
plt.show()


```

```python
# Filter just 'Crash Urgent' reports
crash_df = df[df['Issue Reported'] == 'Crash Urgent']

# Group by month and count
monthly_crashes = crash_df.groupby('Month').size()

monthly_crashes.plot(kind='bar', color='crimson', edgecolor='black')
plt.title("Crash Urgent Reports by Month")
plt.xlabel("Month")
plt.ylabel("Number of Reports")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

```

```python
daily_counts = df.groupby(df['Published Date'].dt.date).size()
daily_counts.plot(figsize=(12, 4), title="Daily Incident Reports Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Reports")
plt.grid(True)
plt.show()

```

```python
# Create a new column for week
df['Week'] = df['Published Date'].dt.to_period('W').apply(lambda r: r.start_time)

# Group by week and count
weekly_counts = df.groupby('Week').size()

# Plot
weekly_counts.plot(figsize=(12, 4), title="Weekly Incident Reports")
plt.xlabel("Week")
plt.ylabel("Number of Reports")
plt.grid(True)
plt.show()


```
